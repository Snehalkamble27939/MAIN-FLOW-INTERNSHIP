## DATA ANALYSIS QUESTIONS

```python
#What is the average score in math (G3)?
average_score= dataframe['G3'].mean()
```

```python
average_score
```

```python
#How many students scored above 15 in their final grade (G3)?
students_scored_above_15 = len(dataframe[dataframe['G3'] > 15])
 
```

```python
#Is there a correlation between study time (study time) and the final grade (G3)?
correlation = dataframe['studytime'].corr(dataframe['G3'])
```

```python
correlation
```

```python
#Which gender has a higher average final grade (G3)?
```

```python
average_grade_by_gender=dataframe.groupby('sex')['G3'].mean()
```

```python
average_grade_by_gender
```

## DATA  VISUALIZATION 

```python
#Plot a histogram of final grades (G3).
plt.figure(figsize=(8,5))
plt.hist(dataframe['G3'], bins=10, edgecolor='black', color='skyblue')
plt.title("Distribution of final grade(G3)")
plt.xlabel('final grade')
plt.ylabel('frequency')
plt.show()
```

```python
#Create a scatter plot between study time (study time) and final grade (G3).
plt.figure(figsize=(8,5))
sns.scatterplot(data=dataframe, x='studytime', y='G3', hue='sex')
plt.title("study time vs final grade")
plt.xlabel('studytime (hours)')
plt.ylabel('final grade')
plt.legend(title='gender')
plt.show()
```

```python
#Create a bar chart comparing the average scores of male and female students
plt.figure(figsize=(8,5))
average_grade_by_gender.plot(kind='bar', color=['blue', 'pink'])
plt.title("average final grade by gender")
plt.ylabel('average final grade')
plt.xlabel('gender')
plt.xticks(rotation=0)
plt.show()
```
